### DS Plotting Functions

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt  # No need for Agg

# Load CSV data
health_data = pd.read_csv("data.csv")

# Plot the data
health_data.plot(x='Average_Pulse', y='Calorie_Burnage', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()  
```
